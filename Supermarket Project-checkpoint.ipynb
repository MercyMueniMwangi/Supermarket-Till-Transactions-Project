# Preparing The Data For (Visualization)

```python
day_dictionary= {2:"Monday", 3:"Tuesday", 4:"Wednesday", 5:"Thursday", 6:"Friday", 7:"Saturday", 1:"Sunday"}

# adding this dictionary as acolumn to our dataframe
dataset['DAYS_OF_WEEK']= dataset['SHOP_WEEKDAY'].map(day_dictionary)
```

```python
#viewing dataset
dataset.head()
```

```python
time_dictionary ={}
for i in range(0,25):
    if i in range(1,12):
        time_dictionary[i] = "Morning"
    if i == 12:
        time_dictionary[i] = "Noon"
    if i in range(13,18):
        time_dictionary[i] = "Afternoon"
    if i in range(18,22):
        time_dictionary[i] = "Evening"
    if i in range(22,0):
        time_dictionary[i] = "Night"
    if i == 0:
        time_dictionary[i] = "Midnight"

# adding this dictionary as acolumn to our dataframe

dataset['TIME_OF_DAY']= dataset['SHOP_HOUR'].map(time_dictionary)
```

# Hypothesis

Null Hypothesis :The mean amount spend is 203 shillings in each STORE_REGION


Alternative Hypothesis: The mean amount is not 203 shillings in each STORE_REGION
    
    H0 : u =203
    H1 : u!= 203
    
    Confidence Interval of 95%
    
    The larger the t score, the more difference there is between groups. The smaller the t score, the more similarity there is between       groups
    
    A t score of 3 means that the groups are three times as different from each other as they are within each other
    
    A large t-score tells you that the groups are different.
    A small t-score tells you that the groups are similar.
    
    
    The calculated t value is then compared to the critical t value from the t distribution table with degrees of freedom df = n - 1 
    and chosen confidence level. If the calculated t value > critical t value, then we reject the null hypothesis.

Here we use T statistics since the population mean and sd are unknown

sample mean = 200.068493

n sample = 73

popultion mean = 203

sd for the sample = 189.783673

```python
import math
numerator =  200.068493 - 203
sqrtn = math.sqrt(73)
Estimated_standarderror_of_the_mean = 189.783673/sqrtn

t_test_calculated = numerator/Estimated_standarderror_of_the_mean
t_test_calculated
```

```python
critical_t_value = 1.994
```

```python
# conclusion
if t_test_calculated>critical_t_value:
    print("REJECT THE NULL HYPOTHESIS \nThe mean amount is not 203 shillings in each STORE_REGION")
else:
    print("FAIL TO REJECT THE NULL HYPOTHESIS \nThe mean amount spend is 203 shillings in each STORE_REGION")
```
